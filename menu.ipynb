## Program Explanation: 

This program is based upon the CW_Preprocessing, Student_Att, Module_Att and Poor_Att programs and it serves the purpose of providing a graphical interface that the end user can easily operate with. 

In order for the menu to be executed as intended the **import_modules_menu ( )** initiliazation must be preformed or a dictionary with the modules names as keys and dataframes being passed under the **neat_table( )** function as values must be provided. 

The graphical interface is devided into 4 sections: 
1. Student Individual Attendance:
   - Here the user is able to input the Student ID for which the program Student_Att will be executed upon pressing the button labeled Get Results.
   
   
2. Module Attendance:
   - The user is able to select from the dropdown menu both a module and a week for which the Module_Att program will be called for.
   - **It should be noted** that the second dropdown menu is linked to the first one i.e weeks options are not fixed, but are activly changing depending on how many weeks the currently selected module has. 
   
   
3. Students with Poor Attendance:
    - The user is able to run the **bad_attendance( )** function from the Poor_Att program by pressing the Poor Att Tables Output button. This will display 2 dataframes. The first one of them will show all the students with poor attendance, who have attended at least a single lecture and the second will display students that have not went to a single lecture. 
    
    - The user is also able to graph the output from above. Using the **graphing ( )** function from the Poor_Att program. Here the users has the option to select what graph to get. It can be inclusive of all students with poor attendance who have attended at least a single lecture or only of the first or last five students in terms of average attendance. This can be done, via choosing an option from the drop down menu and pressing the Poor Att Graph Output. 
    
    
4. Clear Output Options:
    - Here the user is prompted with 2 options. He can either clear all of the outputs generated manually via pressing the Clear All Results button. Alternativly the user can tick the checkbox, which will ensure that after each subsequent entry the previous result is cleared. 
    

And lastly there is the Quit button that closes the program and prints a goodbye message.

## Functions

```python

def clear_if_selected():
    """
    
    Based on whether the checkbox
    in the menu is ticked off, 
    the function will clear output.
    
    """
    if var.get() == 0:
        pass
    else:    
        clear_output()

def clicked_student_individual_att():
    """
    
    Given an user input in the tkinter entry,
    the student_attendance ( ) function is 
    called for the specific entry.
    
    After a specific period the entry box is 
    cleared out. 
    
    The function initially checks whether or not
    it should clear the output, based on the 
    tkinter check button.
    
    """
    clear_if_selected()
    
    user_input = entry_student_individual_att.get()
    student_attendance(modules, user_input)
    entry_student_individual_att.after(2000, 
        lambda: entry_student_individual_att.delete(0, "end"))

    
def pick_week(event, func):
    """
    
    Set the week selection dropdown menu 
    within the Module Att section to 
    have the appropiate weeks for each 
    module. 
    
    """
    for i in func:
        if module_combo.get() == i:
            week_combo.config(value = week_getter(func)[i])
    
def output_module_att():
    """
    
    Gets dataframe and week from 
    the dropdown menus in tkinter 
    and utilizes them in the 
    module_attendance function.
    
    After a certain period it refreshes
    the dropdown menus to their default
    text. 
    
    The function initially checks whether or not
    it should clear the output, based on the 
    tkinter check button.
    
    """
    clear_if_selected()

    entry_df = module_combo.get()
    entry_week = week_combo.get()
    
    module_combo.after(1500, 
        lambda: module_combo.set("Choose Module"))
    week_combo.after(1500, 
        lambda: week_combo.set("Choose Week"))
    
    module_attendance(modules, entry_df, entry_week)


def grapher_output():
    """
    
    Based on the selected option from the 
    respecitve tkinter menu in the 
    Students with Poor Attendance section
    the function returns the requested graph. 
    
    After a certain period it refreshes
    the dropdown menu to its default text. 
    
    The function initially checks whether or not
    it should clear the output, based on the 
    tkinter check button.
    
    """
    clear_if_selected()
    
    if combo_graph.get() == "All students": 
        print("All students with poor attendance: ")
        graphing(modules, bad_attendance(modules))
    elif combo_graph.get() == "Only first five":
        print("First five students in ascending" + 
              " order of average attendance" + 
              " with poor overall attendance: ")
        graphing(modules, bad_attendance(modules).head())
    elif combo_graph.get() == "Only last five":
        print("Last five students in ascending" +
              " order of average attendance" +
              " with poor overall attendance: ")
        graphing(modules, bad_attendance(modules)\
                 [bad_attendance(modules)["Average Attendance %"] !=0]\
                 .tail())
    else: 
        print("Please choose a valid option from the list")
        
    combo_graph.after(2000, 
        lambda: combo_graph.set("Choose Display"))
                                        

def clicked_poor_att():
    """
    
    Returns two dataframes.
    One of them is including students with 
    poor attendance, which have average 
    attendance above 0 %.
    
    The other includes students with poor
    attendance, which have average attendance
    of 0%.
    
    The function initially checks whether or not
    it should clear the output, based on the 
    tkinter check button.
    
    """
    clear_if_selected()
    
    print("Students that have attended at" + 
          " least one of the teaching weeks: ")
    display(bad_attendance(modules)\
            [bad_attendance(modules)["Average Attendance %"] != 0]\
           .style.apply(highlighter, axis = 1)\
           .set_precision(2))
    
    print(" ")
    
    not_str = "\033[1m" + "not" + "\033[0m"
    print("Students that have " + not_str +
          " attended any of the weeks: ")
    display(bad_attendance(modules)\
            [bad_attendance(modules)["Average Attendance %"] == 0]\
           .style.apply(highlighter, axis = 1)\
           .set_precision(2))
    
    print(" ")

    
def close():
    """
    Destroys the tkinter window and 
    prints out a goodbye message as 
    output. 
    
    """
    window.destroy()
    print('Bye!')
```

## Menu

```python
window = Tk()
window.title("Attendance Tracking System") 
window.geometry("450x750")

# Section 1 
task1 = Label(window, 
              text = "Student Individual Attendance\n")
task1.configure(font = ("Ariel", 20, "bold"))
task1.pack()

student_id = Label(window, 
                   text = "Enter student ID: ")
student_id.pack()

entry_student_individual_att = Entry(window, 
                                     width = 10)
entry_student_individual_att.pack()

output_button_individual_att = Button(window, 
                                      text = "Get Results", 
                                      command = clicked_student_individual_att)
output_button_individual_att.pack()

# Section 2 
task2 = Label(window, text = "\nModule Attendance\n")
task2.configure(font = ("Ariel", 20, "bold"))
task2.pack()

# dropdown box
module_combo = ttk.Combobox(window, 
                        value = list(modules.keys()))
module_combo.set("Choose Module")
module_combo.pack()
module_combo.bind("<<ComboboxSelected>>", 
              lambda event:pick_week(event, modules))

# linked box
week_combo = ttk.Combobox(window, 
                          value = [" "])
week_combo.set("Choose Week")
week_combo.pack()
week_combo.bind("<<ComboboxSelected>>", 
                lambda event:pick_week(event, modules))

button_print = Button(window, 
                      text = "Get Results", 
                      command = output_module_att)
button_print.pack()

# Section 3 
task3 = Label(window, 
              text = "\nStudents with Poor Attendance\n")
task3.configure(font = ("Ariel", 20, "bold"))
task3.pack()

output_poor_att = Button(window, 
                         text = "Poor Att Tables Output", 
                         command = clicked_poor_att)
output_poor_att.pack()

clear_space = Label(window, 
                    text = "")
clear_space.pack()

options = ["All students", "Only first five", "Only last five"]
combo_graph = ttk.Combobox(window, value = options)
combo_graph.set("Choose Display")
combo_graph.pack()

button = Button(window, 
                text = 'Poor Att Graph Output', 
                command = grapher_output)
button.pack()

# Clear options
clear_label = Label(window, 
                    text = "\nClear output options\n")
clear_label.configure(font = ("Ariel", 20, "bold"))
clear_label.pack()

var = IntVar()
check = Checkbutton(window, 
                    text = "  I would like to clear\noutput after each entry", 
                    variable = var) 
check.pack()

clear_space = Label(window, 
                    text = "")
clear_space.pack()

clear_button = Button(window, 
                      text = "Clear All Results", 
                      command = clear_output)
clear_button.pack()


# Quit 
end_space = Label(window, 
                  text = "\n")
end_space.pack()
    
close_button = Button(window, 
                      text = "Quit", 
                      command = close)
close_button.pack()

window.mainloop()
```
